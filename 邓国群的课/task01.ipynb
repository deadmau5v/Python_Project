```python
# 读取
df = pandas.read_excel("student1.xlsx")

# （1）对语文英语列缺失的行进行删除
df.dropna(subset=["语文", "英语"], inplace=True)

# （2）对学号列缺失的行进行线性插补
df["学号"] = df["学号"].interpolate(method="linear")

# （3）对数学列缺失的值进行填充，用向前填充法
df["数学"] = df["数学"].fillna(method="ffill")

# （4）对总分列缺失的行进行填充，通过计算语文数学英语的和填充
df["总分"] = df["语文"] + df["数学"] + df["英语"]

# 打印
print(df)
```
